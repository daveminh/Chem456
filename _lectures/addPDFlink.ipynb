```python
import glob
```

```python
FNs = glob.glob('*.md')
for FN in FNs:
  F = open(FN,'r')
  dat = F.read()
  F.close()
  
  lines = dat.split('\n')
  for line_num in range(len(lines)):
    line = lines[line_num]
    if line.startswith('slides_html') or line.startswith('exercise_html'):
      new_line = line.replace('_html','_pdf').replace('/index.html','.pdf')
      print new_line
      lines.insert(line_num+1, new_line)
  lines = [line for line in lines if not line.startswith('#')]
  print '\n'.join(lines)

  F = open(FN,'w')
  F.write('\n'.join(lines))
  F.close()

```
